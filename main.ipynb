For easy for future use, I will make a function for all of the above.

```python
def anime_reccomender_dataset():
    anime = pd.read_csv('C:\\Users\\16195\\Documents\\UCSD\\Kaggles\\Anime Recommender\\data\\anime.csv')
    anime_synopsis = pd.read_csv('C:\\Users\\16195\\Documents\\UCSD\\Kaggles\\Anime Recommender\\data\\anime_with_synopsis.csv')
    
    # Remove animes with less than 1000 watching and completed
    anime_more1000 = anime[anime['Watching'] + anime['Completed']>1000]
    
    # Remove the score columns as we have average score
    anime_clean = anime_more1000.drop(
        columns=[
            'Score-10', 'Score-9', 'Score-8', 'Score-7', 'Score-6', 'Score-5', 'Score-4', 'Score-3', 'Score-2', 'Score-1'
        ]
    )
    
    # Merge anime and anime_synopsis to get the synopsis for each anime, even if the synopsis is empty
    anime_synopsis_merged = anime_synopsis.merge(anime, on='MAL_ID', how='right')
    # replace na with ''
    anime_synopsis_merged['sypnopsis'] = anime_synopsis_merged['sypnopsis'].fillna('')


    # # Merge anime_clean and anime_synopsis
    anime_clean = anime_clean.merge(anime_synopsis_merged[['MAL_ID', 'sypnopsis']], on='MAL_ID')
    
    return anime_clean
```

The dataset is now clean to be used for the recommender system.
